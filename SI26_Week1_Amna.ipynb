# • What is OCR (Optical Character Recognition)?

# **OCR (Optical Character Recognition)** is a technology that converts text from scanned documents, photos, or images into editable and searchable digital text. Instead of typing everything by hand, OCR automatically recognizes the letters and words in an image. It is commonly used to digitize books, forms, receipts, and printed documents, making them easier to store, search, and edit.

# • Why is Urdu OCR harder than English OCR?

# **Urdu OCR** is more difficult than English OCR because Urdu is written in a connected script, where the shape of a letter can change depending on its position in a word. Many Urdu letters also look very similar and are only different because of the number or placement of dots. In addition, Urdu is written from right to left, which makes text recognition more challenging than English.

# • What are 2 real-world situations where Urdu OCR would be useful?

# **Real-World Situations**
# • Digitizing old books and newspapers: Urdu OCR can convert printed books, newspapers, and historical documents into editable digital text, making them easier to preserve, search, and share.
# • Processing official documents: Government offices, schools, hospitals, and businesses can use Urdu OCR to scan printed forms, records, and certificates so they can be stored, edited, and managed digitally without retyping everything.
